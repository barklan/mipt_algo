```python
a = list(map(int, input().split()))
Min = 1000000000000
MinC = 1000000000000
key = 1
for j in range(len(a)):
	for i in range(j, len(a)):
		if a[i] < Min:
			Min = a[i]
			ind = i
	if Min == a[j]:
		key = 0
	Min = MinC
	temp = a[j]
	a[j] = a[ind]
	a[ind] = temp
	if key == 1:
		for i in range(len(a)):
			print(a[i], end = ' ')
		print()
	key = 1
```

<html>

  <p>
   Названия месяцев английские, с заглавной буквы.
  </p>
  <p>
   Диапазон времени 00:00 - 23:59.
  </p>
  <h3>
   Формат входных данных
  </h3>
  <p>
   В первой строке число N - количество записей.
В последующих N строках даты в формате "ЧислоМесяца НазваниеМесяца Год Час:Минута".
Для записи минут используется два знака, т.е. с лидирующим нулём, если количество минут меньше 10.
  </p>
  <h3>
   Формат выходных данных
  </h3>
  <p>
   N строк - отсортированные записи.
  </p>
  <h3>
   Примеры
  </h3>
  <table class="example">
   <thead>
    <tr>
     <td>
      Ввод
     </td>
     <td>
      Вывод
     </td>
    </tr>
   </thead>
   <tbody>
    <tr>
     <td>
      <pre>3
20 December 2019 16:00
20 January 2015 16:00
20 November 2019 16:00</pre>
     </td>
     <td>
      <pre>20 January 2015 16:00
20 November 2019 16:00
20 December 2019 16:00</pre>
     </td>
    </tr>
   </tbody>
  </table>
  <br/>
 </body>
</html>

```python
N = int(input())
a = [0]*N
for i in range(N):
    a[i] = list(map(str, input().split(" ")))
    a[i][0] = int(a[i][0])
    a[i][2] = int(a[i][2])
    temparr = list(map(int, a[i][3].split(":")))
    a[i][3] = temparr[0]
    a[i].append(temparr[1])

for j in range(N*2):
	for i in range(N-1):
		if a[i+1][2] < a[i][2]:
                    a[i], a[i+1] = a[i+1], a[i]

keys = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
for j in range(N*2):
	for i in range(N-1):
            if a[i][2] == a[i+1][2]:
                for k in range(12):
                    if a[i][1] == keys[k]:
                       tmp1 = k
                for k in range(11):
                    if a[i+1][1] == keys[k]:
                       tmp2 = k
                if tmp2 < tmp1:
                    a[i], a[i+1] = a[i+1], a[i]

for j in range(N*2):
	for i in range(N-1):
            if a[i][1] == a[i+1][1] and a[i][2] == a[i+1][2]:
                if a[i+1][0] < a[i][0]:
                            a[i], a[i+1] = a[i+1], a[i]

for j in range(N*2):
	for i in range(N-1):
            if a[i][0] == a[i+1][0] and a[i][1] == a[i+1][1] and a[i][2] == a[i+1][2]:
                if a[i+1][3] < a[i][3]:
                            a[i], a[i+1] = a[i+1], a[i]

for j in range(N*2):
	for i in range(N-1):
            if a[i][3] == a[i+1][3] and a[i][0] == a[i+1][0] and a[i][1] == a[i+1][1] and a[i][2] == a[i+1][2]:
                if a[i+1][4] < a[i][4]:
                            a[i], a[i+1] = a[i+1], a[i]

for i in range(N):
    if a[i][4] < 10:
        tstr = "0" + str(a[i][4])
    else:
        tstr = str(a[i][4])
    a[i][3] = str(a[i][3]) + ":" + tstr
    a[i].pop(4)
    print(*a[i])

```
