```python
m, n = map(int, (input().split(' ')))

if n * 9 < m or m < n:
    print('impossible')
else:
    res = ""
    k = 0
    digit = 9
    while k < n:
        if m - digit > n - k - 2:
            k += 1
            m -= digit
            res = str(digit) + res
            digit = 9
        else:
            digit -= 1
    print(int(res))
```

# F
<html>
 <body>
  <p>
   Вам необходимо определить количество конечных нулей факториала n!. Например, 5! = 120 - число конечных нулей равно 1; 13! = 6227020800 - число конечных нулей равно 2. Будьте осторожны: запись факториала 1000! уже содержит 2568 цифр...
  </p>
  <h3>
   Формат входных данных
  </h3>
  <p>
   Натуральное n
  </p>
  <h3>
   Формат выходных данных
  </h3>
  <p>
   Количество конечных нулей факториала n!
  </p>
  <h3>
   Примеры
  </h3>
  <table class="example">
   <thead>
    <tr>
     <td>

  <br/>
 </body>
</html>

```python
def count (x):
    i = 5
    zeros = 0
    while x >= i:
        zeros += x // i
        i *= 5
    return zeros

a = int(input())
print(count(a))
```
