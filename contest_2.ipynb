```python
N = int(input())
k1 = 0
k2 = 0
k3 = 0
k4 = 0
for i in range(N):
	x, y = input().split()
	x = int(x)
	y = int(y)
	if (x != 0) and (y != 0):
		if (x > 0) and (y > 0):
			k1 += 1
		if (x < 0) and (y > 0):
			k2 += 1
		if (x < 0) and (y < 0):
			k3 += 1
		if (x > 0) and (y < 0):
			k4 += 1
M = max(k1, k2, k3, k4)
if (k1 == M):
	print(1, k1)
if (k2 == M) and (k1 != M):
	print(2, k2)
if (k3 == M) and (k2 != M) and (k1 != M):
	print(3, k3)
if (k4 == M) and (k3 != M) and (k2 != M) and (k1 != M):
	print(4, k4)
```

```python
N = 0
a = 1
S = 0
while (1+1==2):
    a = int(input())
    if a == 0:
        break
    if (N == 0):
        Max = a
        Min = a
    if a > Max:
    	Max = a
    if a < Min:
        Min = a
    S = S + a
    N += 1
Minx = Min - 1
Snat = (Max*(Max+1))/2 - (Minx*(Minx+1))/2
miss = int(Snat - S)
print(miss)
```
