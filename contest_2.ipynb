<html>
 <body>
  <p>
   Дан список точек плоскости с целочисленными координатами.
  </p>
  <p>
   Необходимо определить номер координатной четверти K, в которой находится больше всего точек.
  </p>
  <p>
   Точки, хотя бы одна из координат которых равна нулю, считаются не принадлежащими ни одной четверти и не рассматриваются.
  </p>
  <p>
   Если одновременно в нескольких четвертях одинаковое и при этом максимальное количество точек, то вывести только первую по номеру.
  </p>
  <p>
   <em>
    Замечание:
   </em>
   в данной задаче можно обойтись без использования массива — использовать отдельные четыре счётчика.
  </p>
  <h3>
   Формат входных данных
  </h3>
  <p>
   В первой строке вводится одно целое положительное число – количество точек N.
  </p>
  <p>
   Каждая из следующих N строк содержит координаты очередной точки – два целых числа (первое – координата x, второе – координата y).
  </p>
  <h3>
   Формат выходных данных
  </h3>
  <p>
   В одной строке выводятся номер координатной четверти K, в которой находится больше всего точек и количество точек в ней.
  </p>
  <h3>
   Примеры
  </h3>
  <table class="example">
   <thead>
    <tr>
     <td>
      Ввод
     </td>
     <td>
      Вывод
     </td>
    </tr>
   </thead>
   <tbody>
    <tr>
     <td>
      <pre>7
-3 4
1 2
1 1
0 4
-2 -3
-6 8
-12 1</pre>
     </td>
     <td>
      <pre>2 3</pre>
     </td>
    </tr>
   </tbody>
  </table>
  <br/>
  <table class="example">
   <thead>
    <tr>
     <td>
      Ввод
     </td>
     <td>
      Вывод
     </td>
    </tr>
   </thead>
   <tbody>
    <tr>
     <td>
      <pre>3
-5 -5
-5 -2
1 1</pre>
     </td>
     <td>
      <pre>3 2</pre>
     </td>
    </tr>
   </tbody>
  </table>
  <br/>
 </body>
</html>

```python
N = int(input())
k1 = 0
k2 = 0
k3 = 0
k4 = 0
for i in range(N):
	x, y = input().split()
	x = int(x)
	y = int(y)
	if (x != 0) and (y != 0):
		if (x > 0) and (y > 0):
			k1 += 1
		if (x < 0) and (y > 0):
			k2 += 1
		if (x < 0) and (y < 0):
			k3 += 1
		if (x > 0) and (y < 0):
			k4 += 1
M = max(k1, k2, k3, k4)
if (k1 == M):
	print(1, k1)
if (k2 == M) and (k1 != M):
	print(2, k2)
if (k3 == M) and (k2 != M) and (k1 != M):
	print(3, k3)
if (k4 == M) and (k3 != M) and (k2 != M) and (k1 != M):
	print(4, k4)
```

```python
N = 0
a = 1
S = 0
while (1+1==2):
    a = int(input())
    if a == 0:
        break
    if (N == 0):
        Max = a
        Min = a
    if a > Max:
    	Max = a
    if a < Min:
        Min = a
    S = S + a
    N += 1
Minx = Min - 1
Snat = (Max*(Max+1))/2 - (Minx*(Minx+1))/2
miss = int(Snat - S)
print(miss)
```
